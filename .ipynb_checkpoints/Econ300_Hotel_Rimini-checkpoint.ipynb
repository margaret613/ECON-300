```python
# Import packages
import numpy as np # v 1.21.5
import sklearn # v 1.0.2
import pandas as pd # v 1.4.4
import ydata_profiling as pp # v 3.6.6
import statsmodels.api as sm # v 0.13.2

# Ploting libraries 
import matplotlib
#matplotlib.use('Qt5Agg')  # Use an appropriate backend like 'Qt5Agg' for GUI display
import matplotlib.pyplot as plt # v 3.5.2
import seaborn as sns # v 0.11.2
# Display any generated plots or visualizations directly in the notebook interface
%matplotlib inline 
```

```python
# Read the excel file
hotel_rimini = pd.read_excel('/Users/margaret06/Documents/GitHub/ECON-300/Rimini Hotels Coding.xlsx', sheet_name=0)
```

```python
# Make a copy
hotels_df = hotel_rimini
```

```python
# Print dataframe
hotels_df.head()
```

```python
# Create dummy variables from the 'location' column
dummy_var_location = pd.get_dummies(hotels_df['location'], prefix='location')

# Concatenate the dummy variables with the original DataFrame
hotels_df = pd.concat([hotels_df, dummy_var_location], axis=1)
```

```python
# Print dataframe
hotels_df
```

```python
# Check for the shape of the datafrane
hotels_df.shape
```

```python
# Summary statistics
hotels_df.describe()
```

```python
pp.ProfileReport(hotels_df)
```

```python
# Group by location and room type, calculate average prices for August and December
seasonal_prices = hotels_df.groupby(['location', 'room_type']).agg({'august_price': 'mean', 'december_price': 'mean'})

# Display the table
print(seasonal_prices)

# Plot line graph
seasonal_prices.plot(kind='line')
plt.xlabel('Location and Room Type')
plt.ylabel('Price ($)')
plt.title('Seasonal Price Variations')
plt.show()
```

```python
# Group by location, calculate average prices for each location
location_prices = hotels_df.groupby('location').agg({'august_price': 'mean', 'december_price': 'mean'})

# Display the table
print(location_prices)

# Plot bar chart
ax = location_prices.plot(kind='bar')
plt.xlabel('Location')
plt.ylabel('Average Price ($)')
plt.title('Location Influence on Pricing')

# Annotate each bar with its value
for p in ax.patches:
    ax.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# Create a correlation matrix
correlation_matrix = hotels_df.corr()
sns.set(rc={'figure.facecolor':'w'})
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
plt.title('Correlation Matrix')

# Save the heatmap as a PNG file
plt.savefig('correlation_heatmap.png', format='png')
plt.show()
```
